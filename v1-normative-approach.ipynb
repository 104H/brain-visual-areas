## Normative Approach to Early Visual Areas

How do we know that a ray of light belongs to an object? The answer is in the Central Limit Theorem.
The colour channels of an object will be less Normally Distributed than a collection of objects.

```python
from PIL.Image import open
import numpy as np
import matplotlib.pyplot as plt
```

```python
def channelHistogram(imgpath, bins=15):
    img = open(imgpath)
    plt.imshow(img)
    
    pixels = np.array(img)
    
    red = []
    blue = []
    green = []
    
    for row in pixels:
        for column in row:
            red.append(column[0])
            green.append(column[1])
            blue.append(column[2])
            
    fig, (ax1, ax2, ax3) = plt.subplots(1, 3, sharey=True, figsize=(15, 5))
    fig.suptitle('Histogram of Red, Blue and Green Channels')
    ax1.hist(red, bins=bins, color='red');
    ax2.hist(green, bins=bins);
    ax3.hist(blue, bins=bins, color='green');
```
